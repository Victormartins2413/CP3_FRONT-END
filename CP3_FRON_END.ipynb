```python
import pandas as pd
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import matplotlib.pyplot as plt

# Carregar o dataset
ts_raw = pd.read_csv('https://raw.githubusercontent.com/abnr/ml-data/main/stocks_ABBV.csv', index_col=[0])
ts = ts_raw.reset_index()[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})
ts['ds'] = pd.to_datetime(ts['ds'])

# Configuração dos modelos com diferentes hiperparâmetros
model_params = {
    'm0': {'changepoint_prior_scale': 0.05, 'yearly_seasonality': False, 'weekly_seasonality': False},
    'm1': {'changepoint_prior_scale': 0.1, 'yearly_seasonality': False, 'weekly_seasonality': False},
    'm2': {'changepoint_prior_scale': 0.2, 'yearly_seasonality': True, 'weekly_seasonality': False},
    'm3': {'changepoint_prior_scale': 0.3, 'yearly_seasonality': False, 'weekly_seasonality': True},
    'm4': {'changepoint_prior_scale': 0.4, 'yearly_seasonality': False, 'weekly_seasonality': True},
    'm5': {'changepoint_prior_scale': 0.2, 'yearly_seasonality': False, 'weekly_seasonality': True, 'monthly_seasonality': True},
    'm6': {'changepoint_prior_scale': 0.05, 'yearly_seasonality': True, 'weekly_seasonality': False},
    'm7': {'changepoint_prior_scale': 0.3, 'yearly_seasonality': True, 'monthly_seasonality': True}
}

# Criação dos modelos
models = {}
for model_name, params in model_params.items():
    model = Prophet(changepoint_prior_scale=params.get('changepoint_prior_scale', 0.05),
                    yearly_seasonality=params.get('yearly_seasonality', False),
                    weekly_seasonality=params.get('weekly_seasonality', False))

    # Adicionar sazonalidade mensal quando especificado
    if params.get('monthly_seasonality'):
        model.add_seasonality(name='monthly', period=30.5, fourier_order=5 if model_name == 'm5' else 3)

    models[model_name] = model.fit(ts)

# Função para calcular MAE
def calculate_mae(model):
    df_cv = cross_validation(model, initial='365 days', period='30 days', horizon='30 days')
    df_p = performance_metrics(df_cv)
    return df_p['mae'].mean()

# Armazenar os resultados em um DataFrame para análise
results = {'Model': [], 'MAE': []}
for model_name, model in models.items():
    mae = calculate_mae(model)
    results['Model'].append(model_name)
    results['MAE'].append(mae)

df_validation = pd.DataFrame(results)
print(df_validation)

# Comparar as curvas de validação cruzada de cada modelo em um gráfico
plt.figure(figsize=(12, 8))
for model_name, model in models.items():
    df_cv = cross_validation(model, initial='365 days', period='30 days', horizon='30 days')
    df_cv['error'] = df_cv['yhat'] - df_cv['y']
    df_cv['error_smoothed'] = df_cv['error'].rolling(window=5).mean()
    plt.plot(df_cv['ds'], df_cv['error_smoothed'], label=model_name)

plt.title('Erro Suavizado (yhat - y) para Modelos de Validação Cruzada')
plt.xlabel('Data')
plt.ylabel('Erro (yhat - y)')
plt.legend()
plt.grid()
plt.show()

# Identificar o melhor modelo e justificar
best_model_index = df_validation['MAE'].idxmin()
best_model = df_validation.iloc[best_model_index]
print(f"O melhor modelo é {best_model['Model']} com MAE de {best_model['MAE']:.4f}.")
print(f"Hiperparâmetros do melhor modelo: {model_params[best_model['Model']]}")
```

Isso é um reflexo de seu baixo erro médio absoluto (MAE), pois o modelo está conseguindo prever os valores com menor desvio em relação ao valor real.

No gráfico, isso se reflete em uma curva que segue os padrões de flutuação maiores ao longo do tempo sem reagir de forma excessiva a oscilações de curto prazo.
